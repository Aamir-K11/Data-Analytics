````python
import pandas as pd
import math
import matplotlib.pyplot as plt
from IPython import display
#Reading the csv file.
covid_19 = pd.read_csv("covid_19_data.csv")
original = covid_19.copy()
#Dropping rows which have missing points.
covid_19.dropna(axis=0, how='any', thresh=None, subset=None, inplace=True)

#Printing the first 10 rows 
print("----------------------------------------First 10 rows of dataset----------------------------------------")
pd.set_option('display.max_rows',60 )
display.display(covid_19.head(10))
print("----------------------------------------Summary of dataset----------------------------------------")
display.display(covid_19.describe())
````

#### Question # 02

1. Rank the countries in the dataset in descending order according to the number of deaths.
2. Figure out the most affected provinces/states in top 3 countries with most number of Deaths.
3. Repeat the first two steps for Recovered and Confirmed cases as well.
4. Plot a bar graph for top 10 states in China and US in accordance to the death cases, and recovered cases.

````python
#Sorting dataset in descending order of the number of deaths. 

#Rank countries in descending order according to number of deaths.
covid_19_sorted = covid_19.sort_values(by='Deaths', ascending=0)
print("----------------------------------------Dataset sorted (According to no.of deaths)----------------------------------------")
display.display(covid_19_sorted.head(10))

#Most affected provinces of top 3 countries.(Number of Deaths)
top_countries_provinces = covid_19_sorted.groupby(["Country/Region","Province/State"])["Deaths"].max().reset_index()
top_countries_provinces = top_countries_provinces.sort_values("Deaths",ascending=False)
top_countries_provinces = top_countries_provinces.drop_duplicates(subset="Country/Region",inplace = False).nlargest(3,"Deaths")
print("----------------------------------------Most affected provinces of top 3 countries (No.of deaths)----------------------------------------")
display.display(top_countries_provinces.head(3))

#Rank countries in descending order according to number of confirmed.
covid_19_sorted = covid_19.sort_values(by='Confirmed', ascending=0)
print("----------------------------------------Dataset sorted (According to no.of confirmed cases)----------------------------------------")
display.display(covid_19_sorted.head(10))

#Most affected provinces of top 3 countries.(Number of Confirmed cases).
top_countries_provinces = covid_19_sorted.groupby(["Country/Region","Province/State"])["Confirmed"].max().reset_index()
top_countries_provinces = top_countries_provinces.sort_values("Confirmed",ascending=False)
top_countries_provinces = top_countries_provinces.drop_duplicates(subset="Country/Region",inplace = False).nlargest(3,"Confirmed")
print("----------------------------------------Most affected provinces of top 3 countries(No. of confirmed cases)----------------------------------------")
display.display(top_countries_provinces.head(3))

#Rank countries in descending order according to number of recovered.
covid_19_sorted = covid_19.sort_values(by='Recovered', ascending=0)
print("----------------------------------------Dataset sorted (According to no.of recovered cases)----------------------------------------")
display.display(covid_19_sorted.head(10))

#Most affected provinces of top 3 countries.(Number of Recovered cases).
top_countries_provinces = covid_19_sorted.groupby(["Country/Region","Province/State"])["Recovered"].max().reset_index()
top_countries_provinces = top_countries_provinces.sort_values("Recovered",ascending=False)
top_countries_provinces = top_countries_provinces.drop_duplicates(subset="Country/Region",inplace = False).nlargest(3,"Recovered")
print("----------------------------------------Most affected provinces of top 3 countries(No. of recovered cases)----------------------------------------")
display.display(top_countries_provinces.head(3))

#Top 10 provinces of China
top_10_China = covid_19_sorted.loc[covid_19_sorted['Country/Region'] == "Mainland China"].sort_values(by="Deaths",ascending= False).drop_duplicates(subset="Province/State",inplace = False).head(10)
print("----------------------------------------Top 10 affected provinces of China----------------------------------------")
display.display(top_10_China)
#Top 10 states of  the US
top_10_US = covid_19_sorted.loc[covid_19_sorted['Country/Region'] == "US"].sort_values(by="Deaths",ascending= False).drop_duplicates(subset="Province/State",inplace = False).head(10)
print("----------------------------------------Top 10 affected states of the US----------------------------------------")
display.display(top_10_US)


ax = top_10_China.plot.bar(x="Province/State", y=["Deaths","Recovered"],figsize=(15,5),logy=True,rot=0)
plt.xlabel("Provinces of China")


ax = top_10_US.plot.bar(x="Province/State", y=["Deaths","Recovered"],figsize=(15,5),logy=True,rot=0)
plt.xlabel("States of the US")
````

#### Question # 06

For this question, we will use the `COVID19_open_line_list.csv` file. 

1. Read the data file and examine the features
2. We will be working with these feature columns: `age, sex, symptoms, city, province, lives_in_Wuhan, and travel_history_location`.
3. Since these are textual values, you need to encode them to integers (maintain an encoding dictionary as well).
    * For example in `symptoms` column, you might assign fever -> 1, pneumonia -> 2, headache -> 3, etc.
    * There are several multiple symptoms as well for a single row, for those you can just sum the assigned integers. For example a row might list these symptons: fever, headache, since fever -> 1 and headache -> 3, you can replace fever, headache with 1+3 = 4 and so on. Start the integer encoding from 1 so that when summing the numbers, you donot get the same number back (e.g. 0+3 = 3).
    * You can do the same for other columns, just to avoid ambiguity, you can encode each column differently. 
    * For example, for the `symptoms` column, you starting assigning numbers from 1, let's say there were 4 symptoms, so the numbers would range from 1 - 4. For the next column, you can use numbers from 5 and so on. You are free to encode as you want. 
    * For ease, you can also use sklearn's LabelEncoder.
4. Once you have integer coded the entire dataset (7 columns), you can compute the correlation between city (let city column be `y` for this case) and every other column.
5. Pick the most 2 correlated fields (using the absolute value of correlations) and make `X`.
6. Use MinMaxScaler to scale X. Note that this will output a `np.array`. Make it a DataFrame again and rename the columns appropriately.
7. Print the head (5 rows) and tail (5 rows) of this dataframe.

````python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import OneHotEncoder

COVID19_open_line_list = pd.read_csv("COVID19_open_line_list.csv")
COVID19_open_line_list.dropna(subset=['age','sex','symptoms','city','province','lives_in_Wuhan','travel_history_location'],inplace=True)
features = COVID19_open_line_list[['age', 'sex', 'symptoms', 'city', 'province','lives_in_Wuhan','travel_history_location']]
features.replace(to_replace="Male",value="male",inplace=True)
features.replace(to_replace=" Female",value="female",inplace=True)
# Categorical boolean mask
categorical_feature_mask = features.dtypes==object

# filter categorical columns using mask and turn it into a list
categorical_cols = features.columns[categorical_feature_mask].tolist()

# instantiate labelencoder object
le = LabelEncoder()

# apply le on categorical feature columns
features[categorical_cols] = features[categorical_cols].apply(lambda col: le.fit_transform(col))
features[categorical_cols]

#Find co-relation
features.corr(method="pearson").abs()
X = features[["sex","province"]]
#City and lives_in_Wuhan are strongly co-related
display.display(X)


scaler = MinMaxScaler()
scaled_df = scaler.fit_transform(X)
scaled_df = pd.DataFrame(scaled_df, columns=['sex', 'province'])
display.display(scaled_df)
````
